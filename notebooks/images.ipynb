```python
%load_ext autoreload
%autoreload 2

from dotenv import load_dotenv
load_dotenv()

from IPython.core.display import Image
import pandas as pd

from hotcities.city_image import CityImage
```

```python
cities = pd.read_csv('../data/cities.csv')

cities.head()
```

```python
for index, row in cities.iterrows():
    city_name = row.loc["name"]
    city_localname = None if pd.isna(row['localname']) else row['localname']
    query = f"{city_localname or city_name} skyline"
    print(query)
    try:
        images = CityImage.search(query, num=10)
        cities.at[index, "image_url"] = images[0].url
        print(images[0].url)
    except Exception as e:
        print(e)
```
